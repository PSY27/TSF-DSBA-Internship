#### 2. Fitting of the model

```python
model = Sequential()

model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')

model.fit(x_train, y_train, epochs=5, batch_size=1, verbose=2)
```

#### 3. Making Predictions

```python
inputs = new_data[len(new_data) - len(valid) - 60:].values
inputs = inputs.reshape(-1,1)
inputs = scaler.transform(inputs)
X_test = []

for i in range(60, inputs.shape[0]):
    X_test.append(inputs[i-60:i,0])
X_test = np.array(X_test)

X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1],1))
closing_price = model.predict(X_test)
closing_price = scaler.inverse_transform(closing_price)

_ = plt.figure(figsize=(50, 20))
train = new_data[0:int(len(dataset) * .80)]
valid = new_data[int(len(dataset) * .80):]
valid['Predictions'] = closing_price
_ = plt.plot(train['Close'], label='training data')
_ = plt.plot(valid[['Close','Predictions']], label='testing data')

plt.legend()
```

## News Headline Analysis

    Using EDA and Sentiment Analysis to predict sentiment of headlines in News and how they affect the stock performance.

### 1. Importing Data

    Data downloaded from https://dataverse.harvard.edu/file.xhtml?persistentId=doi:10.7910/DVN/DPQMQH/P2Z4PM&version=1.0

```python
data = pd.read_csv('india-news-headlines.csv')
data.head()
```

### 2. Preprocessing of the Data

```python
only_city = data[data['headline_category'].str.contains('^city\.[a-z]+$', regex=True)]
city = pd.DataFrame(columns = ['city_name'])
city['city_name'] = only_city.headline_category.str.split('.', expand=True)[1]
data = pd.concat([data,city],axis=1)
data.head()
```

```python
data['publish_date'] = data['publish_date']//10000
data.columns.values[0] = 'publish_year'
data.head()
```

```python
data['category'] = data['headline_category'].str.split('.').map(lambda x : x[0])
data.head()
```

```python
data = data[data['category'] != 'unknown']
data.head()
```

### 3. Exploratory Data Analysis

```python
city_df = data.groupby(['city_name']).agg({'headline_text':'count'})
city_df.columns.values[0] = 'headline_count'
city_df = city_df.sort_values(by='headline_count',ascending=False)
city_df.head()
```

```python
_ = plt.figure(figsize=(20,12))
_ = sns.barplot(x=city_df.index, y=city_df['headline_count'])
_ = plt.xticks(rotation = 90)
_ = plt.xlabel('City Name', size=20)
_ = plt.ylabel('Number of articles', size=20)
_ = plt.title('Number of articles per city', size=25)
plt.show()
```

```python
categories = data.groupby(['category']).agg({'headline_text':'count'}).sort_values(by='headline_text',ascending = False)
categories.columns.values[0] = 'count'
categories.reset_index(inplace=True)
categories = categories.head(20)
categories.head()
```

```python
_ = plt.figure(figsize=(20,12))
_ = sns.barplot(x=categories['category'], y=categories['count'])
_ = plt.xticks(rotation = 90)
_ = plt.xlabel('Category', size=20)
_ = plt.ylabel('Number of articles', size=20)
_ = plt.title('Top Category', size=25)
plt.show()
```

### 4. Sentiment Analysis using Text Blob

```python
def polarity(headline):
    result = TextBlob(headline)
    
    if result.sentiment.polarity < 0:
        return 'negative'
    elif result.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'positive'
```

```python
data['Result'] = np.array([polarity(headline) for headline in data['headline_text']])
data.head()
```

```python
sentiment = data.groupby(['Result']).agg({'headline_text':'count'})
sentiment.columns.values[0] = 'count'
sentiment.reset_index(inplace=True)
sentiment
```

```python
_ = plt.figure(figsize=(7,7))
_ = plt.pie(sentiment['count'], labels = sentiment['Result'].unique(), autopct = '%1.1f%%')
_ = plt.title('Sentiment Distribution', size=20)
plt.legend()
plt.show()
```
